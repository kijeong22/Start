# Git Guidebook

## 1. Github 계정 생성
- https://github.com/ 들어가서 계정 등록

### Git 사전 지식
1. commit : 파일을 추가하거나 변경 내용을 저장소에 저장하는 작업
2. push : 파일을 추가하거나 변경 내용을 원격 저장소에 업로드하는 작업
3. 로컬 저장소 & 원격 저장소 : 저장소는 파일이나 디렉토리를 저장하는 장소, 주로 로컬 저장소에서 작업을 수행 후 원격 저장소(Github)에 저장
4. branch : 병렬로 수행되는 여러 버전 관리를 위한 기능

## 2. Git 설치 (window 기준)
1. https://git-scm.com/downloads 여기에 들어가서 'windows' 클릭 후 'Click here to download' 클릭
2. Next 계속 누른 후 Install
3. 설치 다 되면 Launch Git Bash 클릭해서 Bash 실행
4. bash 창에 아래와 같은 명령어를 이용해서 사용자 이름과 이메일 등록
- git config --global user.name "사용자이름"
- git config --global user.email "사용자이메일"
5. 사용자 등록 후 아래 명령어를 이용해서 등록이 잘 되었는지 확인
- git config --list
6. cmd 창에 git 입력해서 잘 나오는지도 확인

## 3. Repository 생성
1. Create New Repository(New) 버튼 클릭
2. Repo name 설정 후 Public으로 설정
3. README 파일 만들거면 'Add a README file'에 체크
4. 나머지는 그냥 그대로 두고 Create repository클릭하면 Repo 생성 완료

## 4. 로컬 저장소 생성 및 원격 저장소와 연결
1. 작업을 할 로컬 저장소를 원하는 경로에 생성
2. cmd에서 해당 경로로 이동 후 "git init"을 통해 현재 디렉토리를 git 저장소로 변경함
3. 아래 명령어를 통해 로컬 저장소를 원격 저장소와 연결 후 확인
- git remote add origin "원격 저장소 URL"
(이 때 origin은 원격 저장소의 default name)
- git remote -v


## 5. 일반적인 작업 과정

0. 아래 과정에 앞서 공동으로 작업을 수행하고 있다면 git pull을 통해 변경사항을 먼저 로컬에 저장한 후 작업을 진행해야 충돌이 발생하지 않음
1. 로컬 저장소에 commit할 파일을 추가
2. 변경 내용을 스테이징 (추가한 파일을 git의 스테이징 영역에 추가)
- git add "파일이름" (파일이름 대신 . 을 사용하면 디렉토리의 모든 파일을 스테이징)
3. 스테이징 영역에 추가한 파일들을 commit으로 생성
- git commit -m "commit message" (commit message는 생략 가능)
4. 저장소에 파일이 추가되었는지 상태 파악
- git status
5. 로컬 저장소의 변경 사항을 원격 저장소에 반영하기 위해 push 실행
- git push origin "branch name"
- 여기서 origin은 원격 저장소의 이름
> 주의사항 : 변경 사항을 스테이징할 때, '.gitignore'파일을 고려해야 함
> gitignore에 기록된 내용은 git이 추적하지 않아야 할 파일들을 정의
> > 아래와 같은 명령어를 우선적으로 사용해서 프로젝트의 git 저장소에서 제외되어야 하는 파일들을 명시적으로 관리
> - git add .gitignore
> - git commit -m "add .gitignore file"

## 6. branch 이용

1. 현재 브랜치 확인
- git branch
2. 새로운 브랜치 생성
- git branch "new branch name"
3. 새로운 브랜치로 이동
- git checkout "new branch name"
- or git switch "new branch name"
4. 위의 일반적인 작업 과정을 모두 진행
- add, commit, push 등
5. 작업이 완료되었으면 해당 브랜치를 다른 브랜치와 병합
- git checkout "병합을 받을 브랜치"
- git merge "병합할 브랜치"
- github에서 "New Pull requests"후 사이트 내에서 merge해주어도 됨
6. 작업을 마친 브랜치 삭제
- 우선 메인 브랜치로 이동 : git checkout main
- 로컬에서 삭제 : git branch -d "브랜치 이름"
- 원격에서 삭제 : git push origin --delete "브랜치 이름"
- 원격 삭제의 경우 github에서도 수행 가능

## 7. 그 외 알아두면 좋은것들


> 추가적으로 자주 사용하는 명령어
- git log : 로컬 저장소의 commit 히스토리를 탐색하는데 사용하는 명령어
- git clone [url] : 기존 원격 저장소를 로컬에 다운로드하기 위하여 사용하는 명령어
- git remote : 원격 저장소의 이름 목록을 표시
- git remote [remote] : 원격 저장소에 대한 자세한 목록보기
- git remote add [remote] [url] : 원격 저장소 추가
- git remote rm [remote] : 원격 저장소 제거
- git remote show [remote] : 원격 저장소에 대한 자세한 정보
- git remote prune [remote] : 해당 원격 저장소의 로컬에서 더 이상 존재하지 않는 원격 브랜치들을 삭제
- git grep "검색 단어" : 특정 단어가 포함된 파일을 검색하고 해당 파일의 어디에 단어가 포함되어 있는지를 확인
- git reset : 로컬 저장소의 commit을 취소하기 위해 사용

> 매번 원격 저장소와 브랜치를 명시적으로 지정하지 않고, git pull, git push 등 간결한 표현으로 작업을 하고싶은 경우
- git branch --set-upstream-to=[remote]/[branch]

> 원격 저장소 변경 방법
- git remote set-url [remote] [새로운 원격 저장소의 URL]
